##### Label encoding is used for the fields with categorical data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
enc.fit(data.International_Plan)
data.International_Plan = enc.transform(data.International_Plan)
```

```python
enc.fit(data.Vmail_Plan)
data.Vmail_Plan = enc.transform(data.Vmail_Plan)
```

```python
enc.fit(data.Churn)
data.Churn = enc.transform(data.Churn)
```

```python
data.head()
```

```python
data.corr()
```

##### The features having the optimal correlation with the output is taken for training

```python
X = data.loc[:,['Vmail_Plan','Day_Mins','Day_Charge','Eve_Mins','Eve_Charge','Night_Mins','International_Mins','International_Charge','CustServ_Calls']]
y = data.Churn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 17)
```

##### As the output variable is imbalanced, SMOTE is used to balance between TRUE and FALSE in the Churn field 

```python
from collections import Counter
from imblearn.over_sampling import SMOTE 
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_sample(X_train, y_train)
print('Resampled dataset shape {}'.format(Counter(y_res)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model =  RandomForestClassifier()
rf_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

predictions = rf_model.predict(X_test)
print(accuracy_score(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

### Model Deployment

```python
from sklearn.externals import joblib
joblib.dump(rf_model, 'churn_model.ml')
```
